```python
# 초기화

# 패턴 추출

pattern = r"\n\d{1,2}시 \d{1,2}분에 영업 시작|\n\d{1,2}시 \d{1,2}분에 영업 시작\n|^\n|^\s|^t"
pattern2 = r'(\n매주 [월화수목금토일]+요일 휴무|\n연중무휴|월\n|\n에 영업 시작|05/29 휴무\n05/29 휴무\n휴무일\n\d{1,2}/\d{1,2}-\d{1,2}/\d{1,2}|\d{1,2}:\d{1,2}에 \n\d{1,2}시 \d{1,2}분에)'
pattern3 = r'(정보없음\n정보없음|현충일|\n휴무일\n\06/\06 현충일|곧 \n|\n\d{1,2}시 \d{1,2}분에 라스트오더|\n매달 \d{1,2}번째 [월화수목금토일]+요일 휴무|\n휴무일\n\d{1,2}/\d{1,2}|\d{1,2}/\d{1,2} 휴무\n\d{1,2}/\d{1,2} 휴무|설명\n)'

row_frm.OPEN_HOUR = row_frm.OPEN_HOUR.str.replace(pattern, '' , regex=True)
row_frm.OPEN_HOUR = row_frm.OPEN_HOUR.str.replace(pattern2, '' , regex=True)
row_frm.OPEN_HOUR = row_frm.OPEN_HOUR.str.replace(pattern3, '' , regex=True)

row_frm.OPEN_HOUR.loc[402] = '영업시간 : 16:00 ~ 03:00'
row_frm.OPEN_HOUR.loc[431] = ''
row_frm.OPEN_HOUR.loc[607] = '영업시간 : 09:00 ~ 19:30'
row_frm.OPEN_HOUR.loc[672] = '영업시간 : 18:00 ~ 23:30 (라스트오더 23:00)'
row_frm.OPEN_HOUR.loc[718] = '영업시간 : 18:00 ~ 23:30 (라스트오더 23:00)'
row_frm.OPEN_HOUR.loc[768] = '영업시간 : 05:30 ~ 15:00'
row_frm.OPEN_HOUR.loc[885] = ''
row_frm.OPEN_HOUR.loc[931] = ''
row_frm.OPEN_HOUR.loc[711] = '18:00에 영업 시작'

def cnull (sent) :
    if sent == '' or sent ==' ' or sent == '/t' :
        sent = None
    else :
        pass
    return sent

row_frm.OPEN_HOUR = row_frm.OPEN_HOUR.apply(cnull)
    
```
