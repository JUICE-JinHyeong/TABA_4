```python
import pandas as pd
```

```python
rest_df = pd.read_excel('제주특별자치도/제주시/식당정보/제주특별자치도_제주시_연동_식당정보_info.xlsx')
rest_df.사업장명.info()
```

```python
rest_name = rest_df.사업장명.drop_duplicates().reset_index(drop=True)
rest_name.info()
```

```python
rest_name_lst = rest_name.tolist()
```

```python
base_frm = pd.read_excel(f'제주특별자치도/제주시/연동/{rest_name_lst[0]}_review.xlsx')
```

```python
columns = base_frm.columns
```

```python
all_rest_frm = pd.DataFrame(columns = columns)
```

```python
all_rest_frm
```

```python
for name in rest_name_lst :
    df = pd.read_excel(f'제주특별자치도/제주시/연동/{name}_review.xlsx')
    all_rest_frm = pd.concat([all_rest_frm , df] , axis = 0)
all_rest_frm = all_rest_frm.reset_index(drop=True)
all_rest_frm
```

```python
# all_rest_frm.columns
# all_rest_frm.My플레이스.astype('str')
```

```python
all_rest_frm.info()
```

```python
all_rest_frm = all_rest_frm.dropna(subset=['리뷰']) # Null 값 제거
all_rest_frm.info()
```

```python
# url이 excel에선 65530개 까지가 제한
# 4분류

all_rest_frm[:20000].to_excel('제주특별자치도/제주시/식당정보/제주특별자치도_제주시_연동_리뷰전체모음_01.xlsx' , index=False)
all_rest_frm[20000:40000].to_excel('제주특별자치도/제주시/식당정보/제주특별자치도_제주시_연동_리뷰전체모음_02.xlsx' , index=False)
all_rest_frm[40000:60000].to_excel('제주특별자치도/제주시/식당정보/제주특별자치도_제주시_연동_리뷰전체모음_03.xlsx' , index=False)
all_rest_frm[60000:].to_excel('제주특별자치도/제주시/식당정보/제주특별자치도_제주시_연동_리뷰전체모음_04.xlsx' , index=False)
```
