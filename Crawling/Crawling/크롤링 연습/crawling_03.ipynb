```python
print('case 01 - bs 이용하여 배우 이름 추출')

url = 'http://www.cine21.com/rank/person/content'
queryString = dict()

queryString['section']      = 'actor'
queryString['period_start'] = '2022-11'
queryString['gender']       = 'all'
queryString['page']         = 1

response = requests.post(url , queryString)
response
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text , 'html.parser')
soup
```

```python
- https://docs.python.org/ko/3/library/re.html#regular-expression-syntax
- \ 다른 문자와 함께 사용되어 특수한 의미로 사용되는 것
- \d , \D == ['^0-9']
- \W['^0-9a-zA-Z'] (역슬래쉬 , W)
- \w == [0-9a-zA-Z]
```

```python
people_lst = soup.select('.people_list > li')
# print('type - ' , type(people_lst))
# print(people_lst)

name_lst_01 = []

for people in people_lst :
    print('배우 이름 : ' , re.sub('\(\w*\)' ,  '' , people.find('div' , {'class' : 'name'}).get_text()))
    name_lst_01.append(re.sub('\(\w*\)' ,  '' , people.find('div' , {'class' : 'name'}).get_text()))
    print('흥행 지수 : ' , people.find('strong').get_text())
    print('출연 영화 : ' , people.select('.mov_list > li span'))
    
    pup = people.find('strong').get_text()
    print(re.sub('[,]' , '' ,pup))
    print('---')
    # 특수문자가 아닌 것들 \W(역슬래쉬 , 대문자 w)
    print(re.sub('[\W]' , '' ,pup))
    print()
```
