```python
import pandas as pd
```

```python
df = pd.read_excel(r'제주도식당/제주특별자치도음식점목록.xlsx')
# 제주특별자치도음식점목록(통합).csv
```

```python
data = df[['사업장명' , '업종구분소분류' , '상세영업상태명' , '소재지전체주소' , '도로명전체주소' , '도로명우편번호']]
data
```

```python
영호 = '푸드트럭/떡카페/다방/아이스크림/과자점/극장/관광호텔/유원지/백화점/이동조리/커피숍(카페랑 중복)/기타 휴게음식점/키즈카페/출장조리'
영호 = 영호.split('/')
영호
```

```python
exclude_elements = ['아이스크림', '과자점', '커피숍(카페랑 중복)']
영민 = [element for element in 영호 if element not in exclude_elements]
영민
```

```python
주혜 = '카페 출장조리 전통찻집 라이브카페 키즈티페 커피숍 편의점 공항 다방 아이스크림 과자점 극장 관광호텔 유원지 백화점'
주혜 = 주혜.split(' ')
주혜
```

```python
혜민 = '출장조리,전통찻집,키즈카페,이동조리,편의점,떡카페,공항,다방,극장,관광호텔,유원지,백화점'
혜민 = 혜민.split(',')
혜민
```

```python
max_length = max(len(영호), len(영민), len(주혜), len(혜민))
data = {
    '영호': 영호 + [None] * (max_length - len(영호)),
    '영민': 영민 + [None] * (max_length - len(영민)),
    '주혜': 주혜 + [None] * (max_length - len(주혜)),
    '혜민': 혜민 + [None] * (max_length - len(혜민))
}
df = pd.DataFrame(data)
df
```

```python
except_rest = ['관광호텔', '극장', '백화점', '유원지', '출장조리' , '편의점']
```

```python
rest_frm = data.query(f'상세영업상태명 == "영업" & 업종구분소분류 != {except_rest} ')
```

```python
rest_frm = rest_frm.dropna().reset_index(drop = True)
```

```python
rest_frm
```

```python
sido = []
sigc = []
sigc_so = []
umdng = []
umdng_so = []
```

```python
for address in rest_frm.소재지전체주소.str.split(' '):
    print(address)
    sido.append(address[0])
    sigc.append(address[0] + ' ' + address[1])
    sigc_so.append(address[1])
    umdng.append(address[0] + ' ' + address[1] + ' ' + address[2])
    umdng_so.append(address[2])
```

```python
dct = {
    '시도' : sido ,
    '시도_시군구' : sigc ,
    '시군구' : sigc_so ,
    '시도_시군구_읍면동' : umdng ,
    '읍면동' : umdng_so
}

address_frm = pd.DataFrame(dct)
address_frm
```

```python
merge_frm = pd.concat([rest_frm,address_frm] , ignore_index=False , axis = 1)
merge_frm
```

```python
# 법정동 코드
# 폐지여부 확인
```

```python
code = pd.read_excel('법정동코드.xlsx')
code = code.query('폐지여부 == "존재"').reset_index(drop=True)
code.법정동코드 = code.법정동코드.astype('str')
```

```python
# 제주도
jeju_code = code.query('법정동명.str.contains("제주")').reset_index(drop=True)
# csv 저장
# jeju_code.to_csv('제주도_법정코드.csv' , index=False , encoding='euc-kr')
jeju_code
```

```python
jeju_code.법정동코드
```

```python
code1 = []
code2 = []
code3 = []

for num in range(len(merge_frm)) :
    for add_num in range(len(jeju_code)) :
        if merge_frm.시도.loc[num] == jeju_code.법정동명.loc[add_num] :
            code1.append(jeju_code.법정동코드.loc[add_num])
        elif merge_frm.시군구.loc[num] == jeju_code.법정동명.loc[add_num] :
            code2.append(jeju_code.법정동코드.loc[add_num])
        elif merge_frm.읍면동.loc[num] == jeju_code.법정동명.loc[add_num] :
            code3.append(jeju_code.법정동코드.loc[add_num])
        else :
            pass
```

```python
code_dict = {'시도코드':code1 , '시군구코드' : code2 , '읍면동코드' : code3}

code_frm = pd.DataFrame(code_dict)
code_frm
```

```python
제주도_음식점 = pd.concat([merge_frm , code_frm] , axis = 1)
제주도_음식점.info()
```

```python
제주도_음식점.to_csv('제주도식당/jeju_rest_addcode.csv' , index=False , encoding='euc-kr')
```
