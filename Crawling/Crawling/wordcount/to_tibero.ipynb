```python
# youngho/nare

import pyodbc

DSNNAME = "Tibero6"
DBUSER = "youngho"
DBPWD = "nare"

cnxn = pyodbc.connect('DSN='+DSNNAME+';UID='+DBUSER+';PWD='+DBPWD)
cursor = cnxn.cursor()
cursor.execute("SELECT REST_NAME from REST_INFO;")

# 1개 row만 가져오기
row = cursor.fetchone()
# 모든 row 가져오기
# row = cursor.fetchall()
while row :
    print(row[0])
    row = cursor.fetchone()
    # row = cursor.fetchall()

cnxn.commit()
cnxn.close()
```
