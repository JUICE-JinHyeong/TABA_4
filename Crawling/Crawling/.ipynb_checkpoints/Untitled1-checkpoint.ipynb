```python
import pandas as pd
```

```python
연동_data_df=pd.read_csv('제주도식당/제주시/연동/id/연동_id_name.csv' , encoding= 'euc-kr')
```

```python
연동_data_df
```

```python
import os

directory_path = '제주도식당/제주시/연동/식당정보(수정본)'  # 디렉토리 경로

if not os.path.exists(directory_path):  # 디렉토리가 존재하지 않으면
    os.makedirs(directory_path)  # 디렉토리 생성
    print('디렉토리가 생성되었습니다.')
else:
    print('디렉토리가 이미 존재합니다.')
    

for name in 연동_data_df.사업장명.tolist() :
    data = pd.read_excel(f'제주도식당/제주시/연동/식당정보/{name}_info.xlsx')
    data['주소id'] = '제주시연동'
    data.to_excel(f'제주도식당/제주시/연동/식당정보(수정본)/{name}_info.xlsx' , index = False)
```
